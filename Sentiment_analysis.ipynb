**Text Cleaning**
1. Sample 10000 rows
2. Remove html tags
3. Remove special character
4. Conmverting every things to lower case
5. Removing stop words
6. Stemming

```python
df=df.sample(10000)
```

```python
df.shape
```

```python
df.info()
```

```python
import re

clean = re.compile('<.*?>')
text = re.sub(clean, '', df.iloc[2]['review'])
```

```python
print(text)
```

```python
# Function to clean html tags

def clean_html(text):
    clean=re.compile('<.*?>')
    return re.sub(clean, '',text)
```

```python


df['review']=df['review'].apply(clean_html)
```

```python
# Converting to lower case

def convert_lower(text):
    return text.lower()



```

```python
df['review']=df['review'].apply(convert_lower)
```

```python
# To remove special character

def remove_special(text):
    x = ''
    for i in text:
        if i.isalnum():
            x = x + i
        else:
            x = x + ' '
    return x
```

```python
df['review']=df['review'].apply(remove_special)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
df
```

```python
stopwords.words("english")
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
# Perform Stemming

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
# Join back
def join_input(list_input):
    return " ".join(list_input)
```

```python
df['review']=df['review'].apply(join_input)
```

```python
df['review']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = df['review']
y = df['sentiment']
```

```python
# Train and test

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
clf1=GaussianNB()
clf2=MultinomialNB()
clf3=BernoulliNB()
```

```python
X=cv.fit_transform(df['review']).toarray()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
y_pred3=clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(f"GaussianNB: {accuracy_score(y_test, y_pred1)}")
print(f"MultinomialNB: {accuracy_score(y_test, y_pred2)}")
print(f"BernoulliNB: {accuracy_score(y_test, y_pred3)}")
```
